```r
library(tidyverse)
library(tidymodels)
library(ggplot2)
```

```r
garbage <- read_csv("garbage.csv") %>%
 mutate(
    Garbage = as_factor(Garbage)) %>%
 mutate(
    Recycling = as_factor(Recycling))
garbage

garbage_recycle <- garbage %>%
    select(Garbage, Recycling)
garbage_recycle

garbage_recycle_plot <- garbage_recycle %>%
  ggplot(aes(Garbage, Recycling)) + 
  geom_bar(position = "dodge", stat="identity") +
 ggtitle("Comparing Garbage vs Recycling") 
garbage_recycle_plot
```
